```python
%matplotlib inline
import matplotlib.pyplot as plt

PSize=[[6],[8],[10],[14],[18]]
PPrice =[[7],[9],[13],[17.5],[18]]

plt.figure()
plt.title("Pizza price plotted against diameter")
plt.xlabel('Diameter in inches')
plt.ylabel('Price in dollars')
plt.plot(PSize,PPrice,'go')
plt.axis([0,25,0,25])

plt.grid(True)
```

```python
from sklearn.model_selection import train_test_split

PSize_train,PSize_test, PPrice_train,PPrice_test = train_test_split(PSize,PPrice,test_size=0.2)#train test train test



print(PSize_train)
print(PSize_test)
print(PPrice_train)
print(PPrice_test)

plt.figure()
plt.title("Pizza price plotted against diameter")
plt.xlabel('Diameter in inches')
plt.ylabel('Price in dollars')
plt.plot(PSize_train,PPrice_train,'go')
plt.plot(PSize_test,PPrice_test,'bo')
plt.axis([0,25,0,25])

plt.grid(True)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(PSize_train,PPrice_train)

print(model.predict(PSize_test))
```

```python
print(model.predict([[18]]))
```

```python
plt.figure()
plt.title("Pizza price plotted against diameter")
plt.xlabel('Diameter in inches')
plt.ylabel('Price in dollars')
plt.plot(PSize_train,PPrice_train,'go')
plt.plot(PSize_test,PPrice_test,'bo')
plt.plot([0,25],[model.predict([[0]])[0][0],model.predict([[25]])[0][0]],color='r')
plt.axis([0,25,0,25])

plt.grid(True)
```

```python
print(model.score(PSize,PPrice))
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(PPrice,model.predict(PSize),multioutput='uniform_average')
print(PPrice)
print("MSE: ",mse)
```
